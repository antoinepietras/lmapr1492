```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.core.operations import SymmOp
import numpy as np
with MPRester("a6NTgQUJ1bGqoLFH") as m:
    structure = m.get_structure_by_material_id("mp-9363")
    w = CifWriter(structure)
    w.write_file('mp9363.cif')
    print("Symétrie 1 : Axe de rotation d'ordre 3 (n°5 du CIF)")
    print('---------------------------------------------------')
    SymmOp_1 = SymmOp.from_xyz_string("-x+y,-x,z")
    print('Matrice associée à la symétrie : -x + y ,-x ,z :')
    print(SymmOp_1.rotation_matrix)
    print("On multiplie cette matrice avec le vecteur colonne suivant pour obtenir notre equation -x + y, -x, z:")
    print(np.array([['x'],['y'],['z']]))
    print(" ")
    print('Symétrie appliquée au Rb situé en (0,1,1) : ')
    print(SymmOp_1.operate((0,1,1)))
    print(' ')
    
    print("Symétrie 2 : Plan miroir 2,-1,0 (n°10 du CIF)")
    print('---------------------------------------------')
    SymmOp_2 = SymmOp.from_xyz_string("-x+y,y,z")
    print('Matrice associée à la symétrie : -x + y, y, z :')
    print(SymmOp_2.rotation_matrix)
    print("On multiplie cette matrice avec le vecteur colonne suivant pour obtenir notre equation -x + y, y, z:")
    print(np.array([['x'],['y'],['z']]))
    print(" ")
    print('Symétrie appliquée au Nd situé en (0,1,0.5) : ')
    print(SymmOp_2.operate((0,1,0.5)))
    print(' ')
    
    print("Symétrie 3 : Rotation hélicoïdale 2_1 (n°21 du CIF)")
    print('---------------------------------------------------')
    SymmOp_3 = SymmOp.from_xyz_string("x-y-1/3,-y+1/3,-z+1/3")
    print('Matrice associée à la symétrie : x -y - 1/3, -y + 1/3, -z + 1/3 :')
    print(SymmOp_3.rotation_matrix)
    print("On multiplie cette matrice avec le vecteur colonne suivant")
    print(np.array([['x'],['y'],['z']]))
    print("Puis on additionne le produit matrice-vecteur avec le vecteur colonne suivant pour obtenir notre equation")
    print("x - y - 1/3, -y + 1/3, -z + 1/3 : ")
    print(np.array([['-1/3'],['1/3'],['1/3']]))
    print(" ")
    print('Symétrie appliquée au S situé en (2/3,1/3,0.1) : ')
    print(SymmOp_3.operate((2/3,1/3,0.1)))
    print(' ')
```
